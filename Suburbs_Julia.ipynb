# Suburbs_Julia

Imports libraries

```julia
using Pkg
using Queryverse
using VegaDatasets, VegaLite
using DataFrames
using CSV
```

Loads CANTERBURY_SUBURB.csv (csv file containing all canterbury suburb SA2 population data) in a new variable called "suburbs".

```julia
suburbs = load("CANTERBURY_SUBURB.csv")
```

Check data type of 'suburbs'

```julia
suburbs |>
  typeof
```

Convert 'suburbs' into a DataFrame type

```julia
suburbs = suburbs |>
    DataFrame
```

Checking 'suburbs' type is a DataFrame

```julia
suburbs |>
  typeof
```

Only keep the June, and years 2013-2021 columns in the suburbs DataFrame.

```julia
suburbs = suburbs[:, [1, 5, 6, 7, 8, 9]]
```

Rename 'June' column to 'Suburb' in the 'suburbs' DataFrame.

```julia
suburbs = rename!(suburbs,:June => :Suburb)
```

Removes the 2013 suburb population column in the suburbs DataFrame.

```julia
suburbs = select!(suburbs, Not(:"2013"))
```

Views suburbs DataFrame

```julia
suburbs
```
